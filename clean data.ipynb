```python
df = pd.read_excel('Raw Data 2024.xlsx')
print(df.head())
```

```python
print(df.isnull().sum())  # นับจำนวน Missing ในแต่ละคอลัมน์
print(df.info())  # ดูโครงสร้างข้อมูล (เช็กค่า NaN)
```

```python
print(df.columns)
```

```python
#Data Preprocessing
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df)
```

```python
df['GENDER'] = df['GENDER'].replace({"M": 0, "F": 1})
df['ADMISSION_TYPE'] = df['ADMISSION_TYPE'].replace({"OPD": 0, "IPD": 1, "Major": 2})
```

```python
# แปลงโรงพยาบาลที่ไม่ระบุ
df['HOSPITAL'] = df['HOSPITAL'].fillna('ไม่ระบุโรงพยาบาล')
```

```python
#ลบแถวที่มีค่า NaN ออก
df = df.dropna(subset=['BENEFIT_AMT'])
display(df)
```

```python
# เลือกช่วงอายุ 20-60 ปี 
df = df[(df['AGE'] >= 20) & (df['AGE'] <= 60)]

display(df)
```

```python
# คำนวณเปอร์เซ็นต์ Missing ของแต่ละคอลัมน์
missing_percent = df.isnull().sum() * 100 / len(df)

# แสดงเฉพาะคอลัมน์ที่มี Missing 
missing_percent = missing_percent[missing_percent > 0]

# เรียงลำดับจากมากไปน้อย
missing_percent = missing_percent.sort_values(ascending=False)

# แสดงผล
print(missing_percent)
```

```python
df.columns = df.columns.str.strip()

# เติมค่าที่หายไปใน CLAIM_TYPE ด้วยค่า mode
df['CLAIM_TYPE'] = df['CLAIM_TYPE'].fillna(df['CLAIM_TYPE'].mode()[0])
df['GENDER'] = df['GENDER'].fillna(df['GENDER'].mode()[0])

# เติมค่าที่หายไปใน GENDER ด้วยค่า mode
df['GENDER'] = df['GENDER'].fillna(df['GENDER'].mode()[0])

# ตรวจสอบว่า Missing 
print(df.isnull().sum())  # ต้องขึ้นเป็น 0 ทุกคอลัมน์
```

```python
#One-Hot Encoding
df_new = pd.get_dummies(df, columns=['PREMIUM_DETAIL'], drop_first=False).fillna(0)

display(df_new)
```

```python
print(df_new.columns)
```

```python
# ตรวจสอบข้อมูลซ้ำ
print(df_new.duplicated().sum())
```

```python
#แปลงให้เป็น ดัมมี่
df_new['PREMIUM_DETAIL_การขับขี่หรือโดยสารรถจักรยานยนต์'] = df_new['PREMIUM_DETAIL_การขับขี่หรือโดยสารรถจักรยานยนต์'].replace({"False": 0, "True": 1}).astype(int)
df_new['PREMIUM_DETAIL_การถูกฆาตกรรมหรือถูกทำร้ายร่างกาย'] = df_new['PREMIUM_DETAIL_การถูกฆาตกรรมหรือถูกทำร้ายร่างกาย'].replace({"False": 0, "True": 1}).astype(int)
df_new['PREMIUM_DETAIL_การทุพพลภาพถาวรสิ้นเชิง'] = df_new['PREMIUM_DETAIL_การทุพพลภาพถาวรสิ้นเชิง'].replace({"False": 0, "True": 1}).astype(int)
df_new['PREMIUM_DETAIL_การสูญเสียอวัยวะ หรือ สายตา'] = df_new['PREMIUM_DETAIL_การสูญเสียอวัยวะ หรือ สายตา'].replace({"False": 0, "True": 1}).astype(int)
df_new['PREMIUM_DETAIL_การเสียชีวิตเนื่องจากอุบัติเหตุ'] = df_new['PREMIUM_DETAIL_การเสียชีวิตเนื่องจากอุบัติเหตุ'].replace({"False": 0, "True": 1}).astype(int)
df_new['PREMIUM_DETAIL_คุ้มครองบัญชีเงินฝากเพิ่มขึ้น 10 เท่า กรณีเสียชีวิตเนื่องจากอุบัติเหตุ สูงสุดไม่เกิน'] = df_new['PREMIUM_DETAIL_คุ้มครองบัญชีเงินฝากเพิ่มขึ้น 10 เท่า กรณีเสียชีวิตเนื่องจากอุบัติเหตุ สูงสุดไม่เกิน'].replace({"False": 0, "True": 1}).astype(int)
df_new['PREMIUM_DETAIL_ค่ารักษาพยาบาล เนื่องจากอุบัติเหตุ (วงเงินความคุ้มครองต่ออุบัติเหตุแต่ละครั้ง)'] = df_new['PREMIUM_DETAIL_ค่ารักษาพยาบาล เนื่องจากอุบัติเหตุ (วงเงินความคุ้มครองต่ออุบัติเหตุแต่ละครั้ง)'].replace({"False": 0, "True": 1}).astype(int)

display(df_new)
```

```python
# ฟังก์ชันสำหรับจัดหมวดหมู่
def categorize_cause(cause):
    if cause in ['อุบัติเหตุ', 'การเจ็บป่วย(IPD)']:
        return 'อุบัติเหตุทั่วไป'
    elif cause in ['ภัยจากยานพาหนะ']:
        return 'อุบัติเหตุจากยานพาหนะ'
    elif cause in ['การขับขี่และโดยสารรถจักรยานยนต์']:
        return 'อุบัติเหตุจากรถจักรยานยนต์'
    elif cause in ['ถูกของมีคมบาด', 'ของแข็งหล่นใส่']:
        return 'อุบัติเหตุจากวัตถุ'
    elif cause in ['ลื่นหกล้ม สะดุดล้ม', 'ตกจากที่สูง', 'ข้อเท้าพลิก/แพลง', 'การเล่นหรือแข่งขันกีฬา']:
        return 'อุบัติเหตุจากการเดินหรือล้ม'
    elif cause in ['ถูกสัตว์/แมลง กัดต่อย']:
        return 'อุบัติเหตุจากสัตว์และแมลง'
    elif cause in ['น้ำร้อนลวก/ไฟลวก']:
        return 'อุบัติเหตุจากความร้อน'
    elif cause in ['การถูกฆ่าและลอบทำร้ายร่างกาย', 'ภัยระเบิด']:
        return 'อุบัติเหตุจากความรุนแรง'
    elif cause in ['สิ่งแปลกปลอมเข้าตา']:
        return 'อุบัติเหตุจากสิ่งแปลกปลอม'
    elif cause in ['ภัยจากเครื่องจักร']:
        return 'อุบัติเหตุจากการทำงาน'
    else:
        return 'อื่น ๆ'

# ใช้ฟังก์ชันเพื่อสร้างหมวดหมู่ใหม่
df_new['CATEGORY_CASE'] = df_new['CAUSE_OF_LOSS'].apply(categorize_cause)

# ดูผลลัพธ์
print(df_new[['CAUSE_OF_LOSS', 'CATEGORY_CASE']].head())

#ลบคอลัมน์ CATEGORY_CASE ออก
df_new = df_new.drop(columns=['CAUSE_OF_LOSS'])

display(df_new)
```
